```python
from scipy import optimize
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

# スカラー関数の最適化
def f(x):
    return x**2 + 10*np.sin(x)

x = np.arange(-10, 10, 0.1)

# 準ニュートン法(BFGS)による最適化
# 勾配法である準ニュートン法では初期値依存で局所解を見つけてしまう
xo = optimize.fmin_bfgs(f, 0)
xo2 = optimize.fmin_bfgs(f, 3)
plt.plot(x, f(x)) 
plt.plot(xo, f(xo),"xr") 
plt.plot(xo2, f(xo2),"xb") 
plt.show()
```

```python
# minimizeの場合はちゃんと大域最適化が求まるパラメータを探してくれる
#どのソルバーが選ばれるかは不明
xo4 = optimize.minimize(f,10)
plt.plot(x, f(x)) 
plt.plot(xo4["x"][0], f(xo4["x"][0]),"xb") 
plt.show()
```
